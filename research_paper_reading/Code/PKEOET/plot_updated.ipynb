```python
import matplotlib.pyplot as plt
import numpy as np

pkeoet_log = 'output/pkeoet_log.txt'
pkeet_log = 'output/pkeet_log.txt'
```

```python
c_num_list = []
pkeoet_enc_list = []
pkeoet_ptest_list = []
pkeoet_total_list = []
pkeoet_dTest_Time = []
pkeoet_dec_Time = []


with open(pkeoet_log, 'r') as f1:
    for line in f1.readlines():
        if line[0] != 'C':
            l = line.split('\t')[:-1]
            c_num_list.append(int(l[0].split('.')[0]))
            pkeoet_enc_list.append(float(l[1]))
            pkeoet_ptest_list.append(float(l[2]))
            pkeoet_total_list.append(float(l[3]))
            pkeoet_dTest_Time.append(float(l[4]))
            pkeoet_dec_Time.append(float(l[5]))

len(c_num_list)
```

```python
# c_num_list = []
pkeet_enc_list = []
pkeet_t_list = []
pkeet_dec_list = []


with open(pkeet_log, 'r') as f3:
    for line in f3.readlines():
        if line[0] != 'C':
            l = line.split('\t')[:-1]
            num = int(l[0].split('.')[0])
            if num in c_num_list:
                pkeet_enc_list.append(float(l[1]))
                pkeet_t_list.append(float(l[2]))
                pkeet_dec_list.append(float(l[3]))


len(pkeet_t_list)
```

Encryption Graph


```python
x = np.array(c_num_list)
y_pkeoet_enc = np.array(pkeoet_enc_list)
y_pkeet_enc = np.array(pkeet_enc_list)


plt.xlabel('Ciphertext Num')
plt.ylabel('Time')

plt.plot(x, y_pkeoet_enc, color='blue', label='PKEOET Encryption Time')
plt.plot(x, y_pkeet_enc, color='red', label='PKEET Encryption Time')

plt.rcParams['figure.dpi'] = 200
plt.title("Encryption")
plt.legend()
plt.show()
```

Decryption Graph

```python
x = np.array(c_num_list)
y_pkeoet_enc = np.array(pkeoet_dec_Time)
y_pkeet_enc = np.array(pkeet_dec_list)


plt.xlabel('Ciphertext Num')
plt.ylabel('Time')

plt.plot(x, y_pkeoet_enc, color='blue', label='PKEOET Decryption Time')
plt.plot(x, y_pkeet_enc, color='red', label='PKEET Decryption Time')

plt.rcParams['figure.dpi'] = 200
plt.title("Decryption")

plt.legend()
plt.show()
```

Total Test Graph

```python
x = np.array(c_num_list)
y_pkeoet_enc = np.array(pkeoet_total_list)
y_pkeet_enc = np.array(pkeet_t_list)


plt.xlabel('Ciphertext Num')
plt.ylabel('Time')

plt.plot(x, y_pkeoet_enc, color='blue', label='PKEOET Test Time')
plt.plot(x, y_pkeet_enc, color='red', label='PKEET Test Time')

plt.rcParams['figure.dpi'] = 200
plt.title("Total Test")

plt.legend()
plt.show()
```

PKEOET Test Times

```python
x = np.array(c_num_list)
y_pkeoet_pTest = np.array(pkeoet_ptest_list)
y_pkeet_dTest = np.array(pkeoet_dTest_Time)
y_pkeet_Total = np.array(pkeoet_total_list)


plt.xlabel('Ciphertext Num')
plt.ylabel('Time')

plt.plot(x, y_pkeoet_pTest, color='blue', label='PKEOET pTest Time')
plt.plot(x, y_pkeet_dTest, color='red', label='PKEOET dTest Time')
plt.plot(x, y_pkeet_Total, color='green', label='PKEOET Test Time')


plt.rcParams['figure.dpi'] = 200
plt.title("PKEOET Tests")

plt.legend()
plt.show()
```
